```python
import duckdb

con = duckdb.connect("../data/trips_data.duckdb")

df = con.execute("""
    SELECT 
        vendor_id, 
        COUNT(*) as total_viagens, 
        AVG(trip_distance) as distancia_media,
        SUM(total_amount) as faturamento_total
    FROM trips_analytics
    GROUP BY 1
""").df()

print(df)
```
